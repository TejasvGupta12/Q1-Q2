```python
import pandas as pd
import numpy as np
from pandasql import sqldf
```

```python
dataset = pd.read_csv('Sample Data.csv')
```

```python
dataset
```

```python
#Converting first row into header
new_header = dataset.iloc[0] 
dataset = dataset[1:] 
dataset.columns = new_header 
```

```python
dataset
```

```python
dataset.reset_index(inplace = True, drop = True) #Reseting index to start from 0
data = dataset.copy() #Copying orginal data to keep a backup
```

```python
# Cleaning names of columns of interest 
# We will use email column to identify unique customers and paid_date column to identify the date they became customer
dataset.rename(columns = {'paid_date\nTEXT':'paid_date','email\nTEXT':'email'}, inplace = True)
```

```python
dataset
```

```python
#Removing time from 'paid_date' column as we are only interested in the date
sliced = dataset.paid_date.str.split()
sliced = sliced.str[0]
dataset.paid_date = sliced
```

```python
#Query to extract month of each entry.
query = sqldf("SELECT strftime('%m', paid_date) Month  FROM dataset ")
query
```

```python
#Query for Cohort Month and Month Number. (Assuming month of interest to be February)
query = sqldf("SELECT CASE strftime('%m', paid_date) WHEN '02' THEN 'February' END 'Cohort Month', paid_date-MIN(paid_date) 'Month Number'  FROM dataset ")
query
```

### QUERY USED ABOVE: "SELECT CASE strftime('%m', paid_date) WHEN '02' THEN 'February' END 'Cohort Month', paid_date-MIN(paid_date) 'Month Number'  FROM dataset "

```python
#Cohort Size
query = sqldf("SELECT strftime('%m', paid_date) Month, COUNT(email) 'Cohort Size' FROM dataset GROUP BY Month HAVING Month='02'")
query
```

### QUERY USED ABOVE: SELECT strftime('%m', paid_date) Month, COUNT(email) 'Cohort Size' FROM dataset GROUP BY Month HAVING Month='02'"

```python
#Retention Users
# Querying a table of users who are common in February and January. (Assuming that first sale was in January 2021)
query_4 = sqldf("SELECT strftime('%m', paid_date) Month, email 'Retained_Users' from dataset where Month=1 INTERSECT SELECT strftime('%m', paid_date) Month, email 'Retained_Users' from dataset where Month=2")

query_4
```

### QUERY USED ABOVE: "SELECT strftime('%m', paid_date) Month, email 'Retained_Users' from dataset where Month=1 INTERSECT SELECT strftime('%m', paid_date) Month, email 'Retained_Users' from dataset where Month=2"
